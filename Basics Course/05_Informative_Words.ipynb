```python
import nltk
```

```python
nltk.corpus.gutenberg.fileids()
```

```python
emma = nltk.corpus.gutenberg.words("austen-emma.txt")
```

```python
emma_fd = nltk.FreqDist(emma)
```

```python
emma_1000_commom_words = [word[0] for word in emma_fd.most_common(1000)]
```

```python
emma_1000_commom_words
```

```python
set(emma) - set(emma_1000_commom_words)
```
